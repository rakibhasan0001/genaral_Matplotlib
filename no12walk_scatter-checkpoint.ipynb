```python
import matplotlib.pyplot as plt


from random import choice
class ramdomwalk:
    
    def __init__(self,num_point=5000):
        self.num_point=num_point
        #walk start from at the point of (0,0)
        self.x_values=[0]
        self.y_values=[0]
    def fill_walk(self):
        while len(self.x_values) < self.num_point:
            
            x_direction=choice([-1,1])
            x_distance=choice([0,1,2,3,4])
            x_step=x_direction*x_distance
            
            #-1 for positive walk 1 for negetive wald
            y_direction=choice([-1,1])
            #chooseing integer number from 0 to 4
            y_distance=choice([0,1,2,3,4])
            y_step=y_direction*y_distance
            
            #reject move that go nowhere
            if x_step ==0 and y_step==0:
                continue
                
            
            #caculate new position
            x=self.x_values[-1]+x_step
            y=self.y_values[-1]+y_step
            
            self.x_values.append(x)
            self.y_values.append(y)
            
rw=ramdomwalk()#create object
rw.fill_walk()
plt.style.use('classic')

fig,ax=plt.subplots()
ax.scatter(rw.x_values,rw.y_values,s=15)
plt.show()



        
            
```

```python
# Linear Search in Python


def linearSearch(array, n, x):

    # Going through array sequencially
    for i in range(0, n):
        if (array[i] == x):
            return i
    return -1


array = [2, 4, 0, 1, 9]
x = 1
n = len(array)
result = linearSearch(array, n, x)
if(result == -1):
    print("Element not found")
else:
    print("Element found at index: ", result)
```
