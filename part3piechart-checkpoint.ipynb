```python
from matplotlib import pyplot as plt

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode=[0,0,0,0.2,0]#python 0.3 means 30% explode in python

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'},explode=explode,shadow=True,
       startangle=40,autopct='%1.1f%%')#wedge properties
plt.show()


#here we learn 
'''startangle
autopct=auto percentage
wedgeprops=a dict %%file
explode
shadow=True/False'''

```
