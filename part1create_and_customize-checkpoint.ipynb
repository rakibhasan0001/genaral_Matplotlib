```python
%matplotlib inline
from matplotlib import pyplot as plt
from ipywidgets import interactive
dev_x=[25,26,27,28,29,30]
dev_y=[20000,30000,40000,5000,60000,70000]
another_y=[34323,78323,78989,32435,32453,45423]

def plotting(c,d):
    plt.plot(c*dev_x,d*dev_y,color='k',linestyle='--',marker='.',label='wikipidia ')
    plt.plot(c*dev_x,d*another_y,'r:o',label='stack flow')#shortcut
    
# here r=red,linestyle=:,marker=o [color][linesytyle][marker]
interactive_plot=interactive(plotting,c=(0,10,1),d=(1,10,1))


plt.show()
```
