```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']



plt.scatter(view_count,likes,c=ratio,alpha=0.80,cmap='summer',edgecolor='blue')
#ratio from csv file
plt.xscale('log')
plt.yscale('log')

cbar=plt.colorbar()
cbar.set_label('like-dislike ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
